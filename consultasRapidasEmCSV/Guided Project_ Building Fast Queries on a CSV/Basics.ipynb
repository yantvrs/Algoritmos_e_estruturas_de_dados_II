```python
import csv
class Inventory():                    
    
    def __init__(self, csv_filename):
        with open(csv_filename) as f: 
            reader = csv.reader(f)
            rows = list(reader)
        self.header = rows[0]        
        self.rows = rows[1:]
        for row in self.rows:              
            row[-1] = int(row[-1])
        self.id_to_row = {}                        
        for row in self.rows:                       
            self.id_to_row[row[0]] = row
        self.prices = set()                          
        for row in self.rows:                        
            self.prices.add(row[-1])
    
    def get_laptop_from_id(self, laptop_id):
        for row in self.rows:                 
            if row[0] == laptop_id:
                return row
        return None   
    
    def get_laptop_from_id_fast(self, laptop_id):  
        if laptop_id in self.id_to_row:           
            return self.id_to_row[laptop_id]
        return None

    def check_promotion_dollars(self, dollars):    
        for row in self.rows:                   
            if row[-1] == dollars:
                return True
        for row1 in self.rows:                  
            for row2 in self.rows:
                if row1[-1] + row2[-1] == dollars:
                    return True
        return False                        
    
    def check_promotion_dollars_fast(self, dollars):
        if dollars in self.prices:
            return True
        for price in self.prices:
            if dollars - price in self.prices:
                return True
        return False
    
    # Esta função recebe um preço mínimo e um preço máximo como argumentos e retorna uma lista de laptops que estão dentro desse intervalo de preços.
    def find_laptops_in_price_range(self, min_price, max_price): 
        laptops_in_range = [] # Inicializa uma lista vazia para armazenar laptops na faixa de preço
        for row in self.rows:
            price = row[-1] # Obtém o preço do laptop na última coluna da linha
            if min_price <= price <= max_price:
                laptops_in_range.append(row) # Adiciona o laptop à lista se estiver dentro da faixa de preço
        return laptops_in_range

    # Esta função recebe uma quantidade mínima de RAM (em gigabytes) e uma quantidade mínima de armazenamento (também em gigabytes) como argumentos e retorna o laptop mais barato que atende a esses critérios de especificações.
    def find_cheapest_laptop_with_specifications(self, min_ram_gb, min_storage_gb):
        matching_laptops = []
        for row in self.rows:
            ram_str = row[7]  # Coluna que contém a informação da RAM (ex: '8GB')
            storage_str = row[8]  # Coluna que contém a informação do armazenamento (ex: '256GB SSD' ou '1TB HDD' ou '512GB')

            # Converter a quantidade de RAM para um número inteiro
            ram_gb = int(ram_str.split('GB')[0])

            # Verificar o tipo de armazenamento (SSD ou HDD)
            if 'SSD' in storage_str:
                storage_type = 'SSD'
            elif 'HDD' in storage_str:
                storage_type = 'HDD'
            else:
                storage_type = None

            # Extrair a capacidade de armazenamento em GB ou TB, conforme apropriado
            storage_capacity = 0
            if 'TB' in storage_str:
                # Extrair a capacidade da string e converter para inteiro (TB para GB)
                storage_capacity_str = storage_str.split('TB')[0]
                if storage_capacity_str.isdigit():
                    storage_capacity = int(storage_capacity_str) * 1000  # Converter de TB para GB (1TB = 1000GB)
            elif 'GB' in storage_str:
                # Extrair a capacidade da string e converter para inteiro (GB)
                storage_capacity_str = storage_str.split('GB')[0]
                if storage_capacity_str.isdigit():
                    storage_capacity = int(storage_capacity_str)

            # Verificar se o laptop atende aos critérios de RAM e armazenamento
            if ram_gb >= min_ram_gb and storage_capacity >= min_storage_gb:
                matching_laptops.append(row)

        # Encontrar o laptop mais barato entre os que atendem aos critérios
        if matching_laptops:
            cheapest_laptop = min(matching_laptops, key=lambda x: x[-1])
            return cheapest_laptop
        else:
            return None

inventory = Inventory('laptops.csv')                 
print(inventory.check_promotion_dollars_fast(1000))  
print(inventory.check_promotion_dollars_fast(442))   
# Encontrar laptops na faixa de preço de $500 a $1000
laptops_in_range = inventory.find_laptops_in_price_range(500, 1000)
print("Laptos com a faixa de preço desejada:")
for laptop in laptops_in_range:
    print(laptop)



# Encontrar o laptop mais barato com pelo menos 8GB de RAM e 256GB de armazenamento
cheapest_laptop = inventory.find_cheapest_laptop_with_specifications(8, 256)
if cheapest_laptop:
    print("Laptop mais barato com as especificações desejadas:")
    print(cheapest_laptop)
else:
    print("Nenhum laptop corresponde às especificações desejadas.")
```

```python
#Vamos analisar a complexidade dos algoritmos implementados em termos de Big O, Big Theta e Big Omega.

#1)  find_laptops_in_price_range:

#Big O (O(n)): A complexidade deste algoritmo é linear, onde "n" representa o número de laptops (ou o tamanho da lista self.rows). Isso ocorre porque ele percorre cada linha da lista uma vez, verificando se o preço está dentro da faixa especificada. No pior caso, ele pode precisar percorrer todas as linhas da lista.

#Big Theta (Θ(n)): A complexidade é linear porque, no melhor e no pior caso, o algoritmo sempre percorre todas as linhas da lista.

#Big Omega (Ω(1)): No melhor caso, se o primeiro laptop na lista estiver dentro da faixa de preço especificada, o algoritmo terminará imediatamente, o que seria uma complexidade constante. No entanto, isso é improvável na prática, e o caso médio e o pior caso têm complexidade linear.

#2) find_cheapest_laptop_with_specifications:

#Big O (O(n)): A complexidade deste algoritmo é linear, onde "n" representa o número de laptops (ou o tamanho da lista self.rows). Isso ocorre porque ele percorre cada linha da lista uma vez, verificando as especificações de RAM e armazenamento.

#Big Theta (Θ(n)): A complexidade é linear porque, no melhor e no pior caso, o algoritmo sempre percorre todas as linhas da lista.

#Big Omega (Ω(1)): No melhor caso, se o primeiro laptop na lista atender às especificações, o algoritmo terminará imediatamente, o que seria uma complexidade constante. No entanto, isso é improvável na prática, e o caso médio e o pior caso têm complexidade linear.

#Ambas as funções têm uma complexidade de tempo linear (O(n)) porque percorrem a lista de laptops uma vez. Elas têm a mesma complexidade de melhor caso (Ω(1)), que é muito raro na prática, e a complexidade média e de pior caso é linear devido à necessidade de percorrer todas as linhas da lista em busca de correspondências.
```
